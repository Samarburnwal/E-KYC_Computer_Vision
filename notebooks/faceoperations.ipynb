```python
import cv2
import matplotlib.pyplot as plt
```

```python
cascade_path = r"haarCascade\haarcascade_frontalface_default.xml"
img_path = r"aadharImages\aadhar.png"
```

```python
img = cv2.imread(img_path)
```

```python
RGBimg = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(RGBimg)
plt.imshow(gray,cmap="gray")
```

```python
face_cascade = cv2.CascadeClassifier(cascade_path)
face_cascade
```

```python
face = face_cascade.detectMultiScale(gray,scaleFactor=1.1,minNeighbors=5)
```

```python
face
#format (x,y,w,h)
[x,y,w,h] = face[0]
type(face)
```

```python
h = int(1.15 * h)
w = int(1.15 * w)
extract_img = img[y:y+h,x:x+w]
```

```python
plt.imshow(extract_img)
```

```python
cv2.imwrite("output.jpg", extract_img)
```
